```python
X = datasets.drop('Price', axis=1)
y = datasets['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X_train, y_train)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
print(regression.get_params())
```

```python
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

```python
plt.scatter(y_test, reg_pred)
```

```python
residual  = y_test - reg_pred
```

```python
sns.displot(residual, kind='kde')
```

```python
plt.scatter(reg_pred, residual)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
reg_pred = regression.predict(X_test)
print('MAE:', mean_absolute_error(y_test, reg_pred))
print('MSE:', mean_squared_error(y_test, reg_pred))
```

```python
from sklearn.metrics import  r2_score
score = r2_score(y_test, reg_pred)
print(score)
```

```python
# Calculate and print the adjusted R-squared
n = len(X_test)
p = X_test.shape[1]
adjusted_r2 = 1 - (1 - score) * (n - 1) / (n - p - 1)
print("Adjusted R-squared:", adjusted_r2)
```

```python
import pandas as pd
import numpy as np

# Fetch the dataset from the original source
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]

# Now you can use 'data' and 'target' for further processing.
# 'data' contains the features and 'target' contains the housing prices.
print(data.shape, target.shape)  # Print the shapes to verify

# Example: Create a Pandas DataFrame
boston_df = pd.DataFrame(data, columns=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT'])
boston_df['Price'] = target  # Add the target variable as 'Price'

# Continue with your analysis using boston_df
boston_df.head()
```

```python
# Assuming 'boston_df' has a column named 'Price' that needs to be removed
X_new = boston_df.drop('Price', axis=1)  # Create a new DataFrame without 'Price' column
# Now you can use X_new for prediction
regression.predict(X_new.iloc[[0]].values)
```

```python
regression.predict(scaler.transform(X_new.iloc[[0]].values))
```
